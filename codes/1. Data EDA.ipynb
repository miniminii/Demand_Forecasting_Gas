```python
data['연월일'] = pd.to_datetime(data['연월일'])
data
```

```python
data['dow'] = data['연월일'].dt.dayofweek
data['doy'] = data['연월일'].dt.dayofyear
data['year'] = data['연월일'].dt.year
data['month'] = data['연월일'].dt.month
data['quarter'] = data['연월일'].dt.quarter
data['weekday'] = data['연월일'].dt.day_name()
data['woy'] = data['연월일'].dt.weekofyear
data['day'] = data['연월일'].dt.day # day of month
data['season'] = data['month'].apply(lambda month_number:(month_number%12 +3)//3)
```

```python
fig = px.line(data,
              x='연월일',
              y='공급량',
              title=f'가스공급량 변화[{min(data.year)} - {max(data.year)}]')
fig.update_traces(line=dict(width=0.3))
fig.update_layout(xaxis_title='Date',
                  yaxis_title='가스공급량')
fig.show()
```

```python
# 연도별 가스공급량 변화
year = data.groupby(['year'],as_index = False).agg({'공급량':'sum'})
```

```python
# plotting
fig = px.line(year, 
              x='year', 
              y='공급량')
fig.update_layout(xaxis_title='YEAR',
                  yaxis_title='가스공급량')
fig.show()
```

2015년 이후 도시가스 수요가수가 증가하는 경향!

## 날짜, 시간 패턴 파악

```python
# 시간별 가스공급량 차이
hour = data.groupby(['시간'],as_index = False).agg({'공급량':'median'})
```

```python
# plotting
fig = px.bar(hour, 
              x='시간', 
              y='공급량', 
              title='시간당 평균가스공급량')
fig.update_layout(xaxis_title='Hour',
                  yaxis_title='가스공급량')
fig.show()
```

```python
# 요일별 시간당 가스공급량
hour_weekday = data.groupby(['시간','weekday'],as_index = False).agg({'공급량':'median'})
```

```python
# plotting
fig = px.line(hour_weekday, 
              x='시간', 
              y='공급량', 
              color='weekday', 
              title='요일별 시간당 평균가스공급량')
fig.update_layout(xaxis_title='Hour',
                  yaxis_title='가스공급량')
fig.show()
```

```python
# 계절별 시간당 가스공급량
hour_season = data.groupby(['시간','season'],as_index = False).agg({'공급량':'median'})
```

```python
hour_season
```

```python
hour_season.loc[hour_season['season'] == 1, 'season'] = '겨울'
hour_season.loc[hour_season['season'] == 2, 'season'] = '봄'
hour_season.loc[hour_season['season'] == 3, 'season'] = '여름'
hour_season.loc[hour_season['season'] == 4, 'season'] = '가을'
```

```python
# plotting
fig = px.line(hour_season, 
              x='시간', 
              y='공급량', 
              color='season', 
              title='계절별 시간당 평균가스공급량')
fig.update_layout(xaxis_title='Hour',
                  yaxis_title='가스공급량')
fig.show()
```

가스공급량 변화 특징  
1) 주말보다 평일에 공급량 증가  
2) 겨울(12~2월)이 다른 계절에 비해 공급량 압도적 많음  
3) 밤 ~ 새벽 시간대에는 공급량 감소, 오전과 저녁 시간대에 크게 증가

가스공급은 겨울>봄>가을>여름  
봄, 가을은 거의 차이 없으나 한 가지 명백한 것은 겨울에 가스 공급량이 모든 공급사에서 크게 늘어난다.

가스 공급 : 가정에 공급하는 도시가스 공급수요 & 발전소로 공급하는 수요  
도시가스 공급수요 : 동고하저의 공급량 변화가 거의 일정  
발전 공급수요 : 당일 전력량의 변화에 따라 매우 불규칙  
-> 도시가스의 공급수요 데이터에 가까움!
> 도시가스 수요는 주로 가정의 취사난방 및 제조공정용으로 사용되는 산업용 수요가 77%

도시가스 사용 : 기온, 요일 패턴, 전일 수요 등과 강한 상관관계를 가지고 있다.
- 가정/상업용 도시가스 소비 증감율은 난방도일(난방수요 크기를 나타내는 지표)의 증감과 거의 유사한 변화 패턴
- 산업용 도시가스 소비  
> 1) 산업업종별 업황, 국가 전체의 경제성장률 영향  
  2) 대체가능한 에너지원과의 상대가격  
  3) 기온변수
